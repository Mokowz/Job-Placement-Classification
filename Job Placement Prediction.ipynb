```python
import pandas as pd
import numpy as np
```

```python
job_df = pd.read_csv("Job_Placement_Data.csv")
job_df
```

There was no null values from the previous analysis. Therefore we will go straight to data preparation

### Column encoding

```python
job_enc = pd.get_dummies(data=job_df, columns=["gender", "specialisation", "ssc_board", "hsc_board", "hsc_subject", "undergrad_degree", "work_experience"])

job_enc
```

```python
job_enc["status"] = job_enc["status"].map({
    "Placed": 1,
    "Not Placed": 0
}) 

job_enc
```

We have mapped the status column(target) such that:

    * Placed = 1
    * Not placed = 0

Now all the data is in numerical values that can be used in training and testing the models.

```python
# Separate the data into features and target
from sklearn.model_selection import train_test_split

np.random.seed(42)

# Create the data
X = job_enc.drop("status", axis=1)
y = job_enc["status"]

# Train test split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3)

X_train
```

```python
y_test
```

### Scaling the numerical values

```python
from sklearn.preprocessing import StandardScaler

# Instantiate the standard scaler
sc = StandardScaler()

# FIt and transform the features
X_train_sc = sc.fit_transform(X_train)
X_test_sc = sc.transform(X_test)
```

### KNN

```python
from sklearn.neighbors import KNeighborsClassifier as KNN

# Instantiate the model
knn = KNN(n_neighbors=7)

# Fit the training data to the model
knn.fit(X_train_sc, y_train)

# Check score
knn.score(X_test_sc, y_test)
```
